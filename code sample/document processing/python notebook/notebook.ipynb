## Streaming Endpoint

```python
import os
import gradio as gr
from dotenv import load_dotenv
import requests
import json
```

```python
load_dotenv()

LLM_ENDPOINT = "https://biz.zygy.com/api/v1/response"
API_KEY = os.getenv("API_KEY")
```

```python
query = input("Enter a query: ")
payload = {
        "querySearch": query,
        "uid": "id",
        "mode":"websearch"
    }



answer_started = False
headers = {"Content-Type": "application/json","Authorization": f"Bearer {API_KEY}"}
with requests.post(LLM_ENDPOINT, json=payload, headers=headers, stream=True, timeout=120) as r:
            r.raise_for_status()
            bot_reply = ""
            
            for chunk in r.iter_lines(decode_unicode=True):
                if not chunk:                    
                    continue


                if chunk.startswith("data:"):
                    chunk = chunk[len("data:"):].strip()
                    chunk = json.loads(chunk)


                    if "content" in chunk['choices'][0]['delta']:
                        chunk = chunk['choices'][0]['delta']['content']

                        if "**Reference**" in chunk:
                            break

                        elif answer_started:
                            print(chunk,end="")
                        

                        elif "Combining results into a final answer..." in chunk:
                            answer_started = True
```

## Generate Template Endpoint

```python
from dotenv import load_dotenv
import requests
import json
```

```python
load_dotenv()

TEMPLATE_ENDPOINT = "https://biz.zygy.com/api/v1/generate"
API_KEY = os.getenv("API_KEY")
```

```python
payload = {
        "templateName": "templateName",
        "uid": "user_id",
        "fileDocs":[{"id":"fileId","type":"xls"}]
}


headers = {"Content-Type": "application/json","Authorization": f"Bearer {API_KEY}"}

with requests.post(TEMPLATE_ENDPOINT, json=payload, headers=headers) as r:
    print(r.json()['report_generated'])
    
```
